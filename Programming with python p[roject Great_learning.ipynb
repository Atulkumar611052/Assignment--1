```python
import math

def get_valid_radius_input():
    while True:
        try:
            radius = float(input("Please enter the radius of the circle: "))
            return radius
        except ValueError:
            print("Invalid input, please enter a number.")

if __name__ == "__main__":
    radius = get_valid_radius_input()
    area = math.pi * radius ** 2
    print("The area of the circle is:", area)

    


```

# Question:10
Convert the ‘InvoiceDate’ into a date format and report the month on which
the maximum sales occur?

```python
clean_data_df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(f"The object type of column 'InvoiceDate' now convert into date format i.e. {df['InvoiceDate'].dtypes}")
```

```python
clean_data_df['month'] = pd.DatetimeIndex(clean_data_df['InvoiceDate']).strftime("%B")

month_max = clean_data_df.groupby('month').sum().sales.sort_values(ascending = False).index[0] 

print(f"The month on which maximum sales occur is {month_max}")
```
